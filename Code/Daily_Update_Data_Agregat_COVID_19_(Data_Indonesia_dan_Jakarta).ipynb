```python
data_id = data_id[['Tanggal','Positif (Indonesia)']]
data_id
```

```python
def make_date(row):
    return datetime(year = int(row.split('-')[2]), 
                    month = int(row.split('-')[1]), 
                    day = int(row.split('-')[0]))

data_id['Tanggal'] = data_id['Tanggal'].apply(make_date)
data_id.set_index(data_id.Tanggal,inplace=True)
data_id.drop(columns=['Tanggal'], inplace=True)

plt.figure(figsize=(8,4))
plt.plot(data_id['Positif (Indonesia)'], 'b')
```

```python
timesteps = 50
```

```python
train = data_id[:len(data_id)-timesteps]['Positif (Indonesia)'].values
test = data_id[len(train):]['Positif (Indonesia)'].values
train = train.reshape(train.shape[0],1)
test = test.reshape(test.shape[0],1)
```

```python
sc = MinMaxScaler(feature_range = (0,1))
train = sc.fit_transform(train)
```

```python
train_X = []
train_y = []

for i in range(timesteps, train.shape[0]):
    train_X.append(train[i-timesteps:i,0])
    train_y.append(train[i,0])

train_X = np.array(train_X)
train_X = train_X.reshape(train_X.shape[0], train_X.shape[1], 1)
train_y = np.array(train_y)
```

```python
print('Training input shape: {}'.format(train_X.shape))
print('Training output shape: {}'.format(train_y.shape))
```

```python
inputs = data_id[len(data_id) - len(test) - timesteps:]
inputs = sc.transform(inputs)

test_X = []

for i in range(timesteps, 100):
    test_X.append(inputs[i-timesteps:i,0])

test_X = np.array(test_X)
test_X = test_X.reshape(test_X.shape[0], test_X.shape[1], 1)
```

```python
test_X.shape
```

```python
model_1 = Sequential() 

model_1.add(SimpleRNN(50, activation='tanh',
                    input_shape=(train_X.shape[1],1), return_sequences = True))
model_1.add(Dropout(0.2))
model_1.add(SimpleRNN(50, activation='tanh', return_sequences = True,))
model_1.add(Dropout(0.2))
model_1.add(SimpleRNN(50, activation='tanh', return_sequences = True,))
model_1.add(Dropout(0.2))
model_1.add(SimpleRNN(50, activation='tanh'))
# output layer to make final predictions 
model_1.add(Dense(1))

model_1.compile(loss='mean_squared_error', optimizer='adam')
model_1.fit(train_X, train_y, epochs=100, batch_size=32, verbose=0)
```

```python
predicted_1 = model_1.predict(test_X)
predicted_1 = sc.inverse_transform(predicted_1)

plt.figure(figsize=(8,4), dpi=80, facecolor='w', edgecolor='k')
plt.plot(test,color="orange",label="Real Confirmed Positive (Indonesia)")
plt.plot(predicted_1,color="c",label="RNN predicted result")
plt.legend()
plt.grid(True)
plt.show()
```

```python
model_2 = Sequential() 
model_2.add(LSTM(70, input_shape=(train_X.shape[1],1)))
model_2.add(Dense(1))

model_2.compile(loss='mean_squared_error', optimizer='adam')
model_2.fit(train_X, train_y, epochs=100, batch_size=32, verbose=0)
```

```python
predicted_2 = model_2.predict(test_X)
predicted_2 = sc.inverse_transform(predicted_2)

plt.figure(figsize=(8,4), dpi=80, facecolor='w', edgecolor='k')
plt.plot(test,color="orange",label="Real Confirmed Positive (Indonesia)")
plt.plot(predicted_2,color="c",label="LSTM Predictions")
plt.legend()
plt.grid(True)
plt.show()
```
